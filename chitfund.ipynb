```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
```

```python
data
```

```python
amount_to_be_pay = len(data.Contribution) * data['Contribution'][0]

paid_amt = amount_to_be_pay - sum(data['Amount returned to everyone in the group'])
```

```python
print("amount_to_be_pay: ", amount_to_be_pay, "\t", "paid_amt: ", paid_amt)
```

```python
data['total_contrib'] = data['Contribution'] - data['Amount returned to everyone in the group']
```

```python
paid_amt = data['total_contrib'].sum()
```

```python
paid_amt
```

```python
def roi(x):
    y = paid_amt
    temp = (x - y) / x * 100
    return temp
```

```python
data['annaualized_return_%'] = data['Net amount recd by Bid winner'].apply(roi) / 2 # divideby 2for annualized roi %
```
